```python
model_names = ['vgg16_conv', 'alexnet_conv']
methods = ['im2col','kn2row','kn2col']
time_execution_4_cores = {}
time_execution_32_cores = {}
```

```python
# 4 cores

import os
import statistics
import re

path = '/teamspace/studios/this_studio/time_execution_4_cores/'
files = os.listdir(path)
for file in files:
    _ = re.split(r'_|\.',file)
    model_name = _[0]
    layer = _[1] + '_' + _[2]
    method = _[3]
    with open(path + file) as f:
        print(file)
        line = f.readline()
        times = line.split()
        times = [float(x) for x in times if float(x) > 0]
        mean_time = statistics.mean(times) / 1000
        if time_execution_4_cores.get(model_name, None) is None:
            time_execution_4_cores[model_name] = {}
        
        if time_execution_4_cores[model_name].get(layer, None) is None:
            time_execution_4_cores[model_name][layer] = {}
        
        time_execution_4_cores[model_name][layer][method] = mean_time
        
```

```python
time_execution_4_cores
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Tên các lớp convolution và các phương pháp
convolution_layers = ['conv1_1', 'conv1_2', 'conv2_1', 'conv2_2', 'conv3_1', 'conv3_2', 'conv4_1', 'conv4_2', 'conv5_1', 'conv5_2']
methods = ['im2col', 'im2row', 'kn2row', 'kn2col']

# Dữ liệu giả lập thời gian thực thi (ms) cho mỗi phương pháp trên mỗi lớp convolution
# Mỗi phần tử trong `execution_times` là thời gian cho một phương pháp tương ứng trên một lớp convolution
execution_times = [
    [time_execution_4_cores['vgg16']['conv1_1'].get(k) for k in methods],  # conv1_1
    [time_execution_4_cores['vgg16']['conv1_2'].get(k) for k in methods],  # conv1_2
    [time_execution_4_cores['vgg16']['conv2_1'].get(k) for k in methods],  # conv2_1
    [time_execution_4_cores['vgg16']['conv2_2'].get(k) for k in methods],   # conv2_2
    [time_execution_4_cores['vgg16']['conv3_1'].get(k) for k in methods],   # conv3_1
    [time_execution_4_cores['vgg16']['conv3_2'].get(k) for k in methods], # conv3_2
    [time_execution_4_cores['vgg16']['conv4_1'].get(k) for k in methods], # conv4_1
    [time_execution_4_cores['vgg16']['conv4_2'].get(k) for k in methods], # conv4_2
    [time_execution_4_cores['vgg16']['conv5_1'].get(k) for k in methods], # conv5_1
    [time_execution_4_cores['vgg16']['conv5_2'].get(k) for k in methods]  # conv5_2
]

# Tạo biểu đồ
x = np.arange(len(convolution_layers))  # Số lượng lớp convolution
width = 0.13  # Độ rộng của mỗi cột

fig, ax = plt.subplots(figsize=(14, 6))

# Vẽ cột cho từng phương pháp, với khoảng cách giữa các phương pháp trong mỗi cụm
for i, method in enumerate(methods):
    times = [layer[i] for layer in execution_times]
    ax.bar(x + i * width - (width * (len(methods) - 1) / 2), times, width, label=method)

# Thêm nhãn và tiêu đề
ax.set_title('Intel Xeon Scalable (Sapphire Rapids) - VGG-16')
ax.set_xlabel('Convolution Layer')
ax.set_ylabel('Execution Time (ms)')
ax.set_xticks(x)
ax.set_xticklabels(convolution_layers)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1))

# Hiển thị biểu đồ
# plt.savefig("exution_time_4_cores.jpg")
plt.show()
```
